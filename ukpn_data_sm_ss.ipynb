```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/souha/OneDrive/Desktop/Masters Thesis/Data/UK Power Networks/ukpn-smart-meter-consumption-substation.csv")
df.head()
```

```python
# List of timestamp columns
timestamp_cols = ['data_collection_log_timestamp', 'insert_time', 'last_modified_time']

# Convert each column to datetime and remove timezone (for naïve datetime)
for col in timestamp_cols:
    df[col] = pd.to_datetime(df[col]).dt.tz_localize(None)

# If you want to keep the timezone info, use:
# for col in timestamp_cols:
#     df[col] = pd.to_datetime(df[col])

df.head()
```

```python
df.shape
```

```python
df.set_index('data_collection_log_timestamp', inplace=True)
df.head()
```

```python
df.sort_index(inplace=True)
df.head()
```

```python
import plotly.express as px

# Calculate frequency counts for the secondary_substation_id column
feeder_counts = df['secondary_substation_id'].value_counts().reset_index()
feeder_counts.columns = ['secondary_substation_id', 'count']

# Create a bar chart using Plotly Express
fig = px.bar(
    feeder_counts,
    x='secondary_substation_id',
    y='count',
    title='Frequency of Secondary substation IDs',
    labels={'secondary_substation_id': 'Secondary substation ID', 'count': 'Count'},
    color_discrete_sequence=['#204251']
)

fig.show(renderer='browser')
```
