```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/souha/OneDrive/Desktop/Masters Thesis/Data/UK Power Networks/ukpn-smart-meter-consumption-lv-feeder.csv")
df.head()
```

```python
# List of timestamp columns
timestamp_cols = ['data_collection_log_timestamp', 'insert_time', 'last_modified_time']

# Convert each column to datetime and remove timezone (for naïve datetime)
for col in timestamp_cols:
    df[col] = pd.to_datetime(df[col]).dt.tz_localize(None)

# If you want to keep the timezone info, use:
# for col in timestamp_cols:
#     df[col] = pd.to_datetime(df[col])

df.head()
```

```python
df.set_index('data_collection_log_timestamp', inplace=True)
df.head()
```

```python
df.sort_index(inplace=True)
df.head()
```

```python
df.describe()
```

```python
import plotly.express as px

# Ensure that lv_feeder_id is treated as a string
df['lv_feeder_id'] = df['lv_feeder_id'].astype(str)

# Calculate frequency counts for the lv_feeder_id column
feeder_counts = df['lv_feeder_id'].value_counts().reset_index()
feeder_counts.columns = ['lv_feeder_id', 'count']

# Create a bar chart using Plotly Express
fig = px.bar(
    feeder_counts,
    x='lv_feeder_id',
    y='count',
    title='Frequency of LV Feeder IDs',
    labels={'lv_feeder_id': 'LV Feeder ID', 'count': 'Count'},
    color_discrete_sequence=['#204251']
)

fig.show(renderer='browser')

```

```python
import plotly.express as px

# Calculate frequency counts for the secondary_substation_id column
feeder_counts = df['secondary_substation_id'].value_counts().reset_index()
feeder_counts.columns = ['secondary_substation_id', 'count']

# Create a bar chart using Plotly Express
fig = px.bar(
    feeder_counts,
    x='secondary_substation_id',
    y='count',
    title='Frequency of Secondary substation IDs',
    labels={'secondary_substation_id': 'Secondary substation ID', 'count': 'Count'},
    color_discrete_sequence=['#204251']
)

fig.show(renderer='browser')
```

```python
# Ensure lv_feeder_id and secondary_substation_id are treated as strings
df['lv_feeder_id'] = df['lv_feeder_id'].astype(str)
df['secondary_substation_id'] = df['secondary_substation_id'].astype(str)

# For each secondary substation, count the number of unique feeder ids and sort descending
substation_feeder_counts = (
    df.groupby('secondary_substation_id')['lv_feeder_id']
      .nunique()
      .reset_index()
      .rename(columns={'lv_feeder_id': 'lv_feeder_id_count'})
      .sort_values(by='lv_feeder_id_count', ascending=False)
)

# For each LV feeder id, count the number of unique secondary substation ids and sort descending
feeder_substation_counts = (
    df.groupby('lv_feeder_id')['secondary_substation_id']
      .nunique()
      .reset_index()
      .rename(columns={'secondary_substation_id': 'secondary_substation_id_count'})
      .sort_values(by='secondary_substation_id_count', ascending=False)
)

print("Number of lv feeder ids for each secondary substation id (Descending):")
print(substation_feeder_counts)

print("\nNumber of secondary substation ids for each lv feeder id (Descending):")
print(feeder_substation_counts)
```

```python
import numpy as np
import plotly.graph_objects as go

# Drop NaN values and ensure the column is numeric
data = pd.to_numeric(df['total_consumption_active_import'], errors='coerce').dropna()

# Aggregate the data into histogram bins (for all customers)
hist, bin_edges = np.histogram(data, bins=100)

# Create a bar chart using the aggregated data and set the marker color to #204251
fig = go.Figure(
    data=[go.Bar(x=bin_edges[:-1], y=hist, width=np.diff(bin_edges), marker_color='#204251')]
)

fig.update_layout(
    title='Histogram of total_consumption_active_import',
    xaxis_title='Value',
    yaxis_title='Frequency'
)

fig.show(renderer='browser')
```

```python
import numpy as np
import plotly.graph_objects as go
import pandas as pd

# Drop NaN values and ensure the column is numeric
data = pd.to_numeric(df['total_consumption_active_import'], errors='coerce').dropna()

# Create a box plot using the cleaned data and set the marker color to #204251
fig = go.Figure(
    data=[go.Box(y=data, marker_color='#204251', boxmean=True)]
)

fig.update_layout(
    title='Box Plot of total_consumption_active_import',
    yaxis_title='Value'
)

fig.show(renderer='browser')
```

```python
import plotly.express as px

# Filter for a specific LV feeder id
selected_id = "83"
df_filtered = df[df['lv_feeder_id'] == selected_id].sort_index()

# Group the data by timestamp (index) and sum the total_consumption_active_import values
df_grouped = df_filtered.groupby(df_filtered.index)['total_consumption_active_import'] \
                        .sum() \
                        .rename_axis('timestamp') \
                        .reset_index()

# Plot the aggregated time series
fig = px.line(
    df_grouped,
    x='timestamp',  
    y='total_consumption_active_import',
    title=f"Time Series of Energy Consumption for LV Feeder ID {selected_id}",
    labels={'timestamp': 'Timestamp', 'total_consumption_active_import': 'Energy Consumption'},
    color_discrete_sequence=['#204251']
)
fig.show(renderer='browser')
```
